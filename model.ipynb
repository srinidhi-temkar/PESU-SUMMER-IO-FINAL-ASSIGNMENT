```python
df.drop(labels = ['Date','RISK_MM','Location','Cloud9am','Cloud3pm','Evaporation','Sunshine'],axis = 1,inplace = True)
df = df.dropna()
```

```python
df['RainToday'].replace({'No': 0, 'Yes': 1},inplace = True)
df['RainTomorrow'].replace({'No': 0, 'Yes': 1},inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
train_df = pd.get_dummies(df,columns=['WindGustDir', 'WindDir3pm', 'WindDir9am'])
train_df.head()
```

```python
x = train_df.drop(labels = ['RainTomorrow'],axis = 1)
y = train_df['RainTomorrow']
x.head()
# y.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
clf = LogisticRegression()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy :',acc)
```
